```python
%pylab inline
from scipy import fftpack
from scipy.fftpack import fft, fftfreq
```

Se lee el archivo y se hace un plot de los puntos corriendolo para que no hayan valores negativos, por practicidad 

```python
data = loadtxt("circulos.dat")
s = 50**-1
x=data[:,0]+3
y=data[:,1]+3
scatter(x,y,s=s)
```

Se hace una malla con celdas cuadradas en las cuales se cuenta cuantos puntos quedan por celda
y se construye una imagen a partir de la cantidad de puntos por celda.

```python
h=500
d=6.0/float(h)
particion=zeros((h,h))
for i in range (0,h):
    for j in range (0,h):
        a=len(x[(x<=d*float(i+1))&(x>d*float(i)) &(y<=d*float(j+1))&(y>d*float(j))])
        particion[i][j]=a

fig=figure(1,figsize = (10,10))
imshow(particion,cmap=cm.Greys_r)
```

Se hace la transformada de fourier de la imagen anterior y se imprime el espectro de potencias
organizado

```python
im_ft = fftpack.fft2(particion)
im_ft_shift = fftpack.fftshift(im_ft)
psd2D= abs(im_ft_shift)**2
fig = figure(1, figsize=(10,10))
imshow(log(psd2D[:,:]),cmap=cm.Greys_r)
```

```python
#pasar el problema a una dimesion
fig = figure(1, figsize=(20,20))
freq1D_sq= abs(im_ft_shift[:,0])**2 + abs(im_ft[:,1])**2
plot (sqrt(im_ft_shift[:,0]**2+im_ft[:,1]**2))
```

```python
a=fftpack.fftshift(sqrt(freq1D_sq))
```

```python
frecuencia=fftshift(fftfreq(h,6.0/h))
plot(a)
```

```python
c=empty(0)
frecuencias_recortadas=empty(0)
for i in range (0,len(frecuencia)):
    if(frecuencia[i]<7)&(frecuencia[i]>-7):
        c=append(c,a[i])
        frecuencias_recortadas=append(frecuencias_recortadas,frecuencia[i])
plot(frecuencias_recortadas,c) 
```

la imagen anterior es el espectro de potencias 1d "radial" de la imagen generada por la malla, pero por el teorema de Wiener-Khinchin esto tambien es la transformada de la autocorelación, la cual será mas o menos periódica y su frecuencia dominante en espacio 

$$
f_{espacial}=\frac{1}{\lambda}
$$
El primer pico va a corresponder a la frecuencia promedio y el segundo a la del primer armónico la cual a su vez va a corresponder a uno sobre el diametro de los círculos pues al desfazar la imagen consigo misma en un diametro es mas probable que se traslapen circulos y la correlación vuelva a ser máxima.
 
Para encontrar la frecuencia a la que corresponde este segundo pico, guiandonos por la gráfica, se buscan espectros de potencia de la imagen en 1d que sean mayores a 500 .
